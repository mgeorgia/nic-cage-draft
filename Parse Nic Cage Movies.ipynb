```python
# Pandas is able to find all instances of a table on a webpage 
# Nic Cage's webpage has tables for each movie and TV show he was in
# Pull the table and write to .csv for later "consumption"
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_html('https://www.rottentomatoes.com/celebrity/nicolas_cage')
```

```python
# should pull 2 different dfs, one of TV shows and one of movies
print(len(df))
```

```python
# Movie dataframe is much longer, use just the movie one
df_lens=[len(x) for x in df]
print(df_lens)
df_cage = df[np.argmax(df_lens)]
df_cage.head()
```

```python
# Rename the columns
df_cage.columns = ['Critics', 'Audience','Title','Credit','Box Office', 'Year']
df_cage.head()
```

```python
#Fix the data types

# Replace "No Score Yet" with 0
df_cage[['Critics', 'Audience']]=df_cage[['Critics', 'Audience']].replace('No Score Yet','0')
# Replace "%" with "" (for some reason df.replace wasn't working with %, used str.replace instead)
df_cage['Critics']=df_cage['Critics'].str.replace('%','')
df_cage['Audience']=df_cage['Audience'].str.replace('%','')
# Convert columns to numeric type
df_cage['Critics'] = pd.to_numeric(df_cage['Critics'])
df_cage['Audience'] = pd.to_numeric(df_cage['Audience'])
df_cage.head()
```

```python
# Save to .csv
df_cage.to_csv('Cage Scores.csv', index=False)
```
